```python
import os 
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir_path = 'penta_corner_detection_data/normal_data/normal_corner20/'
img_path = 'penta_corner_detection_data/normal_data/normal_1/'

a = os.listdir(dir_path)
b = os.listdir(img_path)

for k in range(len(a)):
    file_name = dir_path + a[k]
    img_name = img_path + b[k]
    
    edges = np.loadtxt(file_name)
    
    plt.scatter(edges[:,0], edges[:,1])
    plt.show()
```
